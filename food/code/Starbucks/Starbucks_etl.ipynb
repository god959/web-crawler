```python
#讀取csv檔
import pandas as pd
df = pd.read_csv('E:\Python 3.7\pyetl\Demodb0103/food\Starbucks1\Starbucks_z1.csv',encoding="utf_8")
```

```python
#創立新欄位
df.columns=['Cal_kcal','Protein_g','Fat_g','SF_g','TF_g','Cho_g'
            ,'Sugar_g','Na_g','Name','Brand','Resource','UpdateTime','Portions']
df
```

```python
#Na_g欄位同除1000(毫克轉克)
df['Na_g'] = df['Na_g'].apply(pd.to_numeric)
df['Na_g'] = df['Na_g']/ 1000
```

```python
#新增欄位
df['BraName'] = df['Brand']+"_" + df['Name']
df
```

```python
#存成csv檔
df.to_csv("E:\Python 3.7\pyetl\Demodb0103/food\Starbucks2\Starbucks_e1.csv" , encoding="utf-8", index=False)
```
